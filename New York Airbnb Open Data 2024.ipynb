```python
import pandas as pd
```

```python
airbnb_data = pd.read_csv("new_york_listings_2024.csv")
```

```python
# Display the first few rows
airbnb_data.head()
```

```python
# Get summary statistics
airbnb_data.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting the distribution of room types
plt.figure(figsize=(10, 6))
sns.countplot(x='room_type', data=airbnb_data)
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the distribution of prices by neighbourhood group
plt.figure(figsize=(12, 8))
sns.boxplot(x='neighbourhood_group', y='price', data=airbnb_data)
plt.title('Price Distribution by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.show()
```

```python
# Get the top 10 most reviewed listings
top_reviewed_listings = airbnb_data.nlargest(10, 'number_of_reviews')
top_reviewed_listings
```

```python
# Plotting average reviews per month by neighbourhood group
plt.figure(figsize=(12, 8))
sns.barplot(x='neighbourhood_group', y='reviews_per_month', data=airbnb_data)
plt.title('Average Reviews per Month by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Reviews per Month')
plt.show()
```

```python
# Scatter plot of bedrooms vs. price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bedrooms', y='price', data=airbnb_data)
plt.title('Bedrooms vs. Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter plot of latitude and longitude
plt.figure(figsize=(12, 10))
sns.scatterplot(x='longitude', y='latitude', hue='neighbourhood_group', data=airbnb_data)
plt.title('Map of Airbnb Listings in New York')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Plotting the distribution of availability
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['availability_365'], bins=30, kde=True)
plt.title('Distribution of Availability in a Year')
plt.xlabel('Availability in a Year (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting host listings count distribution
plt.figure(figsize=(10, 6))
sns.histplot(airbnb_data['calculated_host_listings_count'], bins=30, kde=True)
plt.title('Distribution of Host Listings Count')
plt.xlabel('Host Listings Count')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting average price by room type
plt.figure(figsize=(12, 8))
sns.barplot(x='room_type', y='price', data=airbnb_data)
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price')
plt.show()
```

```python
# Plotting room type distribution in each neighbourhood group
plt.figure(figsize=(14, 8))
sns.countplot(x='neighbourhood_group', hue='room_type', data=airbnb_data)
plt.title('Room Type Distribution in Each Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.legend(title='Room Type')
plt.show()
```

```python
# Scatter plot of reviews vs. price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='price', data=airbnb_data)
plt.title('Reviews vs. Price')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter plot of reviews vs. ratings
plt.figure(figsize=(10, 6))
sns.scatterplot(x='number_of_reviews', y='rating', data=airbnb_data)
plt.title('Correlation Between Reviews and Ratings')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.show()
```
